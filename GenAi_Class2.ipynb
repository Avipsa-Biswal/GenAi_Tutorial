```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Activation
import matplotlib.pyplot as plt
import pandas as pd

# Load data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data()

# Extract feature
feature_set = {'average_number_of_rooms': 5}

# Create dataframes
x_train_df = x_train[:, feature_set['average_number_of_rooms']]
x_test_df = x_test[:, feature_set['average_number_of_rooms']]

# Plot data
plt.figure(figsize=(17,7))
plt.scatter(x_train_df, y_train, color='red')
plt.scatter(x_test_df, y_test, color='blue')
plt.xlabel('Average number of rooms')
plt.ylabel('Price')
plt.show()

# Create model
model = Sequential()
model.add(Dense(1, input_shape=(1,)))

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train model
history = model.fit(x_train_df, y_train, epochs=100, batch_size=15, verbose=1)

# Plot loss
plt.plot(history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```
